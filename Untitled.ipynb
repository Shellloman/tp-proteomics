```python
df = df[
    np.logical_and.reduce((
        df['Corrected Abundance ratio (1.53)']!= '#VALEUR!',
        df['Log2 Corrected Abundance Ratio']!= '#VALEUR!',
        df['Abundance Ratio Adj. P-Value: (127. T3 Tc WT) / (126. T0 WT)']!= '#VALEUR!',
        df['-LOG10 Adj.P-val']!= '#VALEUR!',
        
    ))
]
df = df.astype({
    'Accession': str,
    'Description': str,
    'Gene Symbol': str,
    'Corrected Abundance ratio (1.53)': np.float32,
    'Log2 Corrected Abundance Ratio': np.float32,
    'Abundance Ratio Adj. P-Value: (127. T3 Tc WT) / (126. T0 WT)': np.float32,
    '-LOG10 Adj.P-val': np.float32,
})
```

```python
sns.set(rc={'figure.figsize':(12,8)})
sns.histplot(df, x='Log2 Corrected Abundance Ratio')
```

```python
mu, sigma = norm.fit(df['Log2 Corrected Abundance Ratio'])
mu, sigma
```

```python
# _ est le vecteur des valeurs d'abondance
fig, ax = plt.subplots()
hist = ax.hist(df['Log2 Corrected Abundance Ratio'], bins=100) # draw histogram
x = np.linspace(min(df['Log2 Corrected Abundance Ratio']), max(df['Log2 Corrected Abundance Ratio']), 100) # generate PDF domain points
dx = hist[1][1] - hist[1][0] # Get single value bar height
scale = len(df['Log2 Corrected Abundance Ratio'])*dx # scale accordingly
ax.plot(x, norm.pdf(x, mu, sigma)*scale) # compute theoritical PDF and draw it
```

```python
df['surabund'] = (df['Log2 Corrected Abundance Ratio'] > mu + sigma) & (df['-LOG10 Adj.P-val'] < 3)
```

```python
sns.set(rc={'figure.figsize':(8, 8)})
sns.scatterplot(data=df, y='-LOG10 Adj.P-val', x='Log2 Corrected Abundance Ratio', hue='surabund')
```
